```python
import pandas as pd
import numpy as np
```

```python
cources=pd.read_csv('courses.csv')
student=pd.read_csv('students.csv')
nov=pd.read_csv('reg-month1.csv')
dec=pd.read_csv('reg-month2.csv')
matches=pd.read_csv('matches.csv')
deliveries=pd.read_csv('deliveries(1).csv')
```

```python
reg=pd.concat([nov,dec],ignore_index=True)
```

```python
reg.head()
```

# inner join

```python
pd.merge(left=student,right=reg,how='inner',on='student_id')
```

# left join

```python
pd.merge(left=cources,right=reg,how='left',on='course_id')
```

# right join

```python
temp_df=pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Azhar','Shyam'],
    'partner':[28,26,27]
})
```

```python
temp_df
```

```python
student=pd.concat([student,temp_df],ignore_index=True)
```

```python
student
```

```python
pd.merge(left=student,right=reg,how='right',on='student_id')
```

# outer join

```python
pd.merge(left=student,right=reg,how='outer',on='student_id')
```
